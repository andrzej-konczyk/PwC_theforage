# The historical repayment percentages, i.e. every repayment’s share of the origination amount

```python
df1 = df.drop(['Unnamed: 0'], axis=1)
percent = round(df1.div(df1['Origination Amount'], axis='index') * 100,2)
percent
```

```python
percent['Origination Amount'] = df['Origination Amount']
percent
```

```python
idx =0
percent.insert(loc=idx, column ='Unnamed: 0', value=df['Unnamed: 0'])
```

```python
percent =percent.rename(columns={'Unnamed: 0': 'Credit date'})
```

# Creation temporary dataframe to check how much prct was already done per each credit

```python
prct_new= percent.drop(columns=['Credit date', 'Origination Amount'], axis=0)
prct_new['prct_sum']=prct_new.sum(axis=1)
prct_new
```

```python
prct_new['remaining'] = percent['Origination Amount']*(100-prct_new['prct_sum'])
remaining = round(prct_new['remaining'].sum(axis=0),2)
remaining
```

```python
print(f'For day 31.12.2020 there is {remaining} remaining CHF')
```
